## Exercise 3:

Write a function to generate sum of ‘n’ numbers of Fibonacci sequence where ‘n’

$$F_{seq} = FibonacciSum(n)$$

where $n$ is the input to the function while $F_{seq}$ is the output of the function
containing the sum of $n$ numbers generated by Fibonacci sequence.

```python
def fibonaccisum(n):
    def fibonacci(m):
        if m == 0 or m == 1:
            return [0] if m == 0 else [0,1]
        result = fibonacci(m-1)
        return result + [result[-1] + result[-2]]
    return sum(fibonacci(n))

fibonaccisum(10)
```

## Exercise 4:

Write a function to generate ‘n’ numbers of Fibonacci sequence and their sum such
that numbers are not divisible by 2 or 5. ‘n’ is the input to the function.

$$F_{sum},F_{seq} = FibonacciFilter(n)$$

where $n$ is the input to the function while $F_{sum}$ and $F_{seq}$ are the outputs of the
function containing the sum and an array or list of $n$ numbers generated by
Fibonacci sequence which are not divisible by 2 or 5.

```python
def fibonaccifilter(n):
    def fibonacci(m):
        if m == 0 or m == 1:
            return [0] if m == 0 else [0,1]
        result = fibonacci(m-1)
        return result + [result[-1] + result[-2]]
    seq = [x for x in fibonacci(n) if x % 2 != 0 and x % 5 != 0]
    return (sum(seq), seq)

fsum, fseq = fibonaccifilter(10)
(fsum, fseq)
        
```

```python
# Is this not the same as Exercise 7? Is it supposed to be the reverse? I'll do the reverse...
def str2int(num):
    if num.strip().lower() in ['ten', 'eleven','twelve']:
        match num.strip().lower():
            case 'ten': return 10
            case 'eleven': return 11
            case 'twelve': return 12
    split_str = num.strip().lower().split()
    result = []
    for clip in split_str:
        for i,x in enumerate(['one','tw','th','fo','fi','si','se','ei','ni']):
            if clip.startswith(x):
                result.append(i + 1)
    if split_str[0].endswith('teen'):
        return int('1' + str(result[0]))
    elif len(result) > 1:
        return int(str(result[0]) + str(result[1]))
    else:
        return result[0]

str2int('ten')
```

## Exercise 6:

Write a function which can add two similar size matrices taking both the matrices
as inputs of the function. The matrices are taken as list of lists.

```python
def matrix_sum(matrix1, matrix2):
    result = []
    if isinstance(matrix1, int):
        return matrix1 + matrix2
    for x,y in zip(matrix1, matrix2):
        result.append(matrix_sum(x, y))
    return result

matrix_sum([[0,1,2],[1,2,0]], [[1,1,1],[3,3,6]])
```

## Exercise 8:

Write a function which takes two list of same size as inputs and return a list
containing alternate items. For example, mergelists([1,2,3],[4,5,6]) will return
[1,4,2,5,3,6].

```python
def mergelists(list1, list2):
    result = []
    for x,y in zip(list1, list2):
        result += [x]
        result += [y]
    return result

mergelists([1,2,3],[4,5,6])
```
